### 마커표시

아이콘은 부트스트랩 아이콘 사용(웹html,css)

````python
m =  folium.Map(location=[35.117479, 129.090394] , zoom_start=14)
folium.Marker(location=[35.117479, 129.090394], popup='부경대 용당캠퍼스',icon=folium.Icon(color='red')).add_to(m)
m
````

### 마커 커스터마이즈
iframe= folium.IFrame('부경대학교<br>용당캠퍼스<br>지역산업맞춤산업단')



````python
m= folium.Map(location=[35.117479,129.090394],zoom_start=14)
iframe= folium.IFrame('부경대학교<br>용당캠퍼스<br>지역산업맞춤산업단')
popup = folium.Popup(iframe, min_width=300, max_width=350)
folium.Marker(location=[35.117479,129.090394] , popup=popup, icon=folium.Icon(color='red',icon='leaf')).add_to(m)

m
````
